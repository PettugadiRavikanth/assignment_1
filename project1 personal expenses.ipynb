```python

categories = ["Investment", "Stationery", "Subscriptions", "Home Essentials","Health","Transport","Shopping","Entertainment","Food","Health"]
payment_mode = ["Cash", "Debit Card", "Credit Card", "Net Banking", "Wallet", "UPI"]
descriptions = ["Investment in Term Insurance",
                "Investment in Mutual Funds",
                "Purchase of stationery",
                "Spotify subscription",
                "YouTube Premium subscription",
                "Investment in Stocks",
                "Quarterly school fees payment",
                "Purchase of home essentials",
                "Fresh fruits and vegetables",
                "Purchase of sports & fitness",
                "Weekly grocery shopping",
                "Investment in Insurance Policy",
                "Doctor visit",
                "Bus fare",
                "Cloths",
                "Groceries",
                "Movie tickets",
                "Streaming service"
]
```

```python

# Function to assign descriptions based on category
def get_description(category):
    if category == "Food":
        return "Groceries"
    elif category == "Investment":
        return random.choice(["Investment in Term Insurence","Investment in Mutual Funds","Investment in Stocks",  "Investment in Insurance Policy"])
    elif category == "Health":
        return "Doctor visit"
    elif category == "Transport":
        return "Bus fare"
    elif category == "Shopping":
        return "Clothes"
    elif category == "Entertainment":
        return "Movie tickets"
    elif category == "Stationery":
        return random.choice(["Office supplies",  "Purchase of stationery"])
    elif category == "Subscriptions":
        return random.choice(["Streaming service",  "Spotify subscription",  "YouTube Premium subscription"])
    return "Miscellaneous"
```

```python
#GENERATE JANUARY DTAES
def generate_jan_dates():
    start_date = datetime(2024, 1, 1)
    end_date = datetime(2024, 1, 30)
    delta = timedelta(days = 1)
    dates = []
    while start_date <= end_date:
        dates.append(start_date.date())
        start_date += delta
    return dates
```

```python
# Generate data
def gen_exp_data_jan(num_entries=120):
    data = []
    jan_dates = generate_jan_dates()
    for i in range(num_entries):
          category = random.choice(categories)
          payment_method = random.choice(payment_mode)
          expense_jan = {
                "Date": jan_dates[ i % len(jan_dates)],
                "Categories": category,
                "Payment_Mode": payment_method,
                "Description": get_description(category),
                "Amount_paid": round(random.uniform(10, 1000), 2),
                "Cash_Back": round(random.uniform(0, 5), 2) if payment_method in ["wallet","UPI"] else 0
          }
          data.append(expense_jan)
    return pd.DataFrame(data)
```

```python
exp_data_jan = gen_exp_data_jan(120)
```
